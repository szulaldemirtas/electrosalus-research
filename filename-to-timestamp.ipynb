```python
import os
import time
```

```python
#Single File

directory_path = '/Users/zulaldemirtas/git/electrosalus-research/records'

# Listing all files in the directory
audio_files = os.listdir(directory_path)
file = 'untitled.wav'

# Full path of the file
file_path = os.path.join(directory_path, file)

if os.path.isfile(file_path):  # Check if it's a regular file
    # Obtaining the creation time (in seconds)
    # of the file/folder (datatype=int)
    print(file_path)

    creation_time = os.path.getmtime(file_path)

    # Converting the time to an epoch string
    # (the output timestamp string would
    # be recognizable by strptime() without
    # format quantifiers)
    t_obj = time.localtime(creation_time)

    # Transforming the time object to a timestamp
    # of ISO 8601 format
    form_t = time.strftime("%Y-%m-%dT%H-%M-%S", t_obj)
    print(form_t)

    # Since colon is an invalid character for a
    # Windows file name Replacing colon with a
    # similar looking symbol found in unicode
    # Modified Letter Colon " " (U+A789)
    form_t = form_t.replace(":", "꞉")


    # Renaming the filename to its timestamp
    os.rename(
        file_path,
        os.path.join(directory_path, form_t + os.path.splitext(file)[1])
    )

else:
    print(f"{file} is not a regular file. Skipping.")
```

```python
import os
import time

directory_path = './timestamps/'

# Listing all files in the directory
audio_files = os.listdir(directory_path)

for file in audio_files.copy():
    # Full path of the file
    file_path = os.path.join(directory_path, file)

    if os.path.isfile(file_path):  # Check if it's a regular file
        # Obtaining the creation time (in seconds)
        # of the file/folder (datatype=int)
        t = os.path.getctime(file_path)

        # Converting the time to an epoch string
        # (the output timestamp string would
        # be recognizable by strptime() without
        # format quantifiers)
        t_str = time.ctime(t)

        # Converting the string to a time object
        t_obj = time.strptime(t_str)

        # Transforming the time object to a timestamp
        # of ISO 8601 format
        form_t = time.strftime("%Y-%m-%dT%H-%M-%S" + str(file), t_obj)

        # Since colon is an invalid character for a
        # Windows file name Replacing colon with a
        # similar looking symbol found in unicode
        # Modified Letter Colon " " (U+A789)
        form_t = form_t.replace(":", "꞉")

        # Renaming the filename to its timestamp
        os.rename(
            file_path,
            os.path.join(directory_path, form_t + os.path.splitext(file)[1])
        )
    else:
        print(f"{file} is not a regular file. Skipping.")
```
